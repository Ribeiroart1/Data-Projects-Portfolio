```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Create a DataFrame with our hypothetical data
data = {'Payment_Method': [1, 2, 1, 2, 1, 2, 1, 2],
        'Price': [50, 60, 55, 65, 48, 58, 51, 63],
        'Region': [1, 2, 1, 2, 1, 2, 1, 2],
        'Customer_Satisfaction': [80, 85, 78, 88, 82, 87, 79, 90]}

df = pd.DataFrame(data)
```

```python
# Trying to fit the multiple linear regression model
model = ols('Customer_Satisfaction ~ Payment_Method + Price + Region', data=df).fit()
```

```python
# Displaying regression results
print(model.summary())
```

```python
# To show Display regression results
print(model.summary())
```

```python
# Perform ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```
